```python
import os
import pandas as pd
```

```python
# ✅ Path to check folders in
path = "./archive/additional_images"

# ✅ Get folder names from file system
folder_list = [name.strip().replace(" ", "") for name in os.listdir(path) if os.path.isdir(os.path.join(path, name))]

# ✅ Another list to compare against
path_to_file='./archive/pokemon_with_folders.csv'
df = pd.read_csv(path_to_file,sep=',')
df["Name"] = df["Name"].str.strip().str.replace(" ", "", regex=False)
reference_list = df["Name"].tolist()

print("Folders from system:", folder_list, "\ncount:", len(folder_list))
print("Reference folders:", reference_list, "\ncount:", len(reference_list))
```

```python
# ✅ Find missing items
folder_list_lower = [f.lower() for f in folder_list]
reference_list_lower = [r.lower() for r in reference_list]

missing_in_folders = [r for r in reference_list if r.lower() not in folder_list_lower]
missing_in_reference = [f for f in folder_list if f.lower() not in reference_list_lower]

print("Missing folders:", missing_in_folders, "\ncount: ", len(missing_in_folders))
print("Extra folders:", missing_in_reference,  "\ncount: ", len(missing_in_reference))
```
