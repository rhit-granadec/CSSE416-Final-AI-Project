```python
import pandas as pd
```

```python
path_to_file='./archive/pokemon_with_folders.csv'
df = pd.read_csv(path_to_file,sep=',')
df = df[["Name", "Type", "Other Type"]]
print(df.shape)
df.head()
```

```python
noEvo = df.rename(columns={'Type': 'Type1', 'Other Type': 'Type2'})
noEvo.head()
```

# Duel Type analyis

```python
iterateTry = noEvo.copy()
pokemonType = []
for pokemon in iterateTry.itertuples(index=False):
    if type(pokemon.Type2) is float:
        currentType = pokemon.Type1
    elif pokemon.Type1 < pokemon.Type2:
        currentType = pokemon.Type1 + '-' + pokemon.Type2
    else:
        currentType = pokemon.Type2 + '-' + pokemon.Type1
    pokemonType.append(currentType)

iterateTry["Type"] = pokemonType
iterateTry = iterateTry.drop(["Type1", "Type2"], axis=1)
iterateTry.head()
```

```python
counts = iterateTry['Type'].value_counts()
print(counts.shape)
print(counts.to_string())
```

# Primary Type Analysis

```python
counts = noEvo['Type1'].value_counts()
print(counts.shape)
print(counts.to_string())
```

```python
baseline = counts.max()/noEvo.shape[0]
print("Our baseline single type predictor is right",baseline*100,"percent of the time")
```

# Both Type Analysis

```python
noEvoManip = noEvo.copy()
noEvoManip = noEvoManip.fillna("NoSecondType")
counts1 = noEvoManip['Type1'].value_counts()
counts2 = noEvoManip['Type2'].value_counts()
totalCount = counts1.add(counts2, fill_value=0)
totalCount = totalCount.sort_values(ascending=False)
print(totalCount.shape)
print(totalCount.to_string())
```

```python
totalDataPoints = noEvoManip.shape[0]
correctGuesses = 0
for pokemon in noEvoManip.itertuples(index=False):
    if pokemon.Type1 == "Water" and pokemon.Type2 == "NoSecondType":
        correctGuesses = correctGuesses + 1

correctPercent = correctGuesses/totalDataPoints
print("Our Baseline Predictor was correct",correctPercent*100,"percent of the time.")
```
